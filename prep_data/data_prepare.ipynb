```python
import pandas as pd
```

处理：尝试对所有的category变量进行独热编码
有可能出现数据维度过高的情况

```python
from sklearn.preprocessing import OneHotEncoder
# 创建OneHotEncoder对象
encoder = OneHotEncoder()
category = ['Status_of_existing_checking_account','Credit_history','Purpose','Savings_account/bonds','Present_employment_since','Other_debtors/guarantors','Property','Other_installment_plans','Housing','Job','Telephone','Foreign_worker']
# 将属性17进行独热编码
encoded_German_data = encoder.fit_transform(German_data[category])

# 将稀疏矩阵转换为密集矩阵
dense_encoded_German_data = encoded_German_data.toarray()

# 将密集矩阵转换为DataFrame
df_encoded_German_data = pd.DataFrame(dense_encoded_German_data, columns=encoder.get_feature_names_out(category))

# 打印DataFrame
print(df_encoded_German_data)

# # 使用标签编码（Label Encoding）
# from sklearn.preprocessing import LabelEncoder

# # 创建LabelEncoder对象
# encoder = LabelEncoder()

# # 将属性17进行标签编码
# encoded_German_data = encoder.fit_transform(German_data[category])

# # 打印标签编码后的特征
# print(encoded_German_data)
```

```python
German_data.columns
German_data.drop("Telephone",axis = 1, inplace = True) # here this telephone do not have effect on our model
numerical_var = ['Duration_in_month','Credit_amount','Present_residence_since','Age_in_years','Number_of_existing_credits_at_this_bank','Number_of_people_being_liable_to_provide_maintenance_for','Good/Bad', 'sex']
numerical = German_data[numerical_var]
```

```python
result = pd.concat([df_encoded_German_data, numerical],axis = 1)
print(result)
```
